```python
import psycopg2
```

```python
def get_cursor_raw(host, database, user, port):
    connection = psycopg2.connect(host=host,
                                  database=database,
                                  user=user,
                                  port=port)
    cursor = connection.cursor()
    return connection, cursor
```

```python
def single_insert(connection, insert_req):
    """ Execute a single INSERT request """
    try:
        cursor.execute(insert_req)
        connection.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        connection.rollback()

        return 1
    
```
